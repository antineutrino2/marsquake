```python
import pandas as pd
import os
import urllib.request
import matplotlib.pyplot as plt
import requests
from io import StringIO

url = "https://service.iris.edu/irisws/mars-event/1/query?format=text&longitude=135.6234&latitude=4.5024&minradius=1"

response = requests.get(url)

plain_text = response.text

# Split lines using the newline character
lines = plain_text.strip().split('\n')

# Delimit values using the pipe character
data = [line.split('|') for line in lines]
```

```python
df = pd.DataFrame(data[1:], columns=data[0])
df
```
